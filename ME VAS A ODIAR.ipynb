Debe porder gestionar cualquier columna
o segmenteos de columnas, y cantidad de filas,

```python
df_copy = df
```

```python
id = 32
columna = 'Dato_3'
dato = 100

try:
    df_copy.loc[[id], [columna]] = dato
except KeyError:
    # En caso de que no exista la fila, creo una nueva
    df_copy = pd.concat([df_copy, 
                         pd.DataFrame({columna:[dato]})], 
                         ignore_index=True)
    
    # Esto lo hago por facilidad, es mas facil respetar la primary key, que ignorarla
    print(f'ADVERTENCIA: Se ha creado una nueva fila con el id {len(df_copy) - 1}.')

df_copy
```
